```python
import requests # requests機能をインポートします。
from bs4 import BeautifulSoup #BeautifulSoupをインポート
from selenium import webdriver #seleniumをインポート
REQUEST_URL = 'https://suumo.jp/jj/chintai/ichiran/FR301FC001/?fw2=&mt=9999999&cn=9999999&ta=13&et=9999999&sc=13103&shkr1=03&ar=030&bs=040&ct=9999999&shkr3=03&shkr2=03&srch_navi=1&mb=0&shkr4=03&cb=0.0' #アクセス先をREQUEST_URLを代入します。
res = requests.get(REQUEST_URL) #リクエストしたデータをresに代入します。
```

```python
res.encoding
```

```python
soup = BeautifulSoup(res.text,"html.parser")
```

```python
soup
```

```python
# html全体を格納しているsoupからsectionというcssセレクタを使って抽出
rental_section = soup.select("cassetteitem")[0]  #[0]は何個目の物件を指定。数字を入れ替えて変更可能
rental_section
```
