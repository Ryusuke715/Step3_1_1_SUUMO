```python
#ライブラリのインポート
from time import sleep #Pythonの実行を●秒間停止させる
from bs4 import BeautifulSoup
import requests
```

```python
#アクセスするURLを設定する
url = "https://suumo.jp/chintai/tokyo/sc_shinagawa/?page={}"

#変数d_listに空のリストを作成する
d_list = []

#スクレイピングする最大ページ数を設定
max_page = 100
```

```python
#1~max_pageをループする
for i in range (1,max_page+1):
    print('d_listの大きさ:',len(d_list))

    #変数target_urlに、アクセス先のURLを格納する
    target_url = url.format(i)

    #print()してtarget_urlを確認する
    print(target_url)

    #Requestsを使って設定したURLにアクセスする
    r = requests.get(target_url)

    #1秒の間隔をあけて実行
    sleep(1)

    #取得したHTMLをBeautifulSoupで解析する
    soup = BeautifulSoup(r.text)

    #すべての物件情報（20件）を取得する
    contents = soup.find_all('div',class_='cassetteitem')

    #各物件情報をforループで取得する
    for content in contents:

        #物件情報と部屋情報を取得しておく
        detail = content.find('div',class_='cassetteitem-detail')
        table = content.find('table',class_='cassetteitem_other')

        #物件情報から必要な情報を取得する
        title = detail.find('div',class_='cassetteitem_content-title').text
        address = detail.find('li',class_='cassetteitem_detail-col1').text
        access1 = detail.find_all('div', class_='cassetteitem_detail-text')[0].text
        access2 = detail.find_all('div', class_='cassetteitem_detail-text')[1].text
        access3 = detail.find_all('div', class_='cassetteitem_detail-text')[2].text
        age = detail.find('li', class_='cassetteitem_detail-col3').find_all('div')[0].text
        height = detail.find('li', class_='cassetteitem_detail-col3').find_all('div')[1].text

        #部屋情報のブロックから、各部屋を取得する
        tr_tags = table.find_all('tr',class_='js-cassette_link')

        #各部屋情報をforループで取得する
        for tr_tag in tr_tags:

            #部屋情報の行から、欲しい情報を取得する
            floor,price,first_fee,capacity = tr_tag.find_all('td')[2:6]

            #さらに細かい情報を取得する
            fee,management_fee = price.find_all('li')
            deposit,gratuity = first_fee.find_all('li')
            madori,menseki =capacity.find_all('li')

            #取得したすべての情報を辞書に格納する
            d = {
                'title':title,
                'address':address,
                'access1':access1,
                'access2':access2,
                'access3':access3,
                'age':age,
                'height':height,
                'floor':floor.text,
                'fee':fee.text,
                'management_fee':management_fee.text,
                'deposit':deposit.text,
                'gratuity':gratuity.text,
                'madori':madori.text,
                'menseki':menseki.text,
            }

            d_list.append(d)
```

```python
#CSVを作成するためpandasをインポート
import pandas as pd
import re
```

```python
#d_listの2番目まで中身を確認してみる
d_list[:2]
```

```python
#変数d_listを使って、データフレームを作成する
df =pd.DataFrame(d_list)
```

```python
df.head()
```

```python
#access列を分割し、余分な文字は削除し、数値に変換
df[['access1_route', 'access1_station', 'access1_time']] = df['access1'].str.extract(r'([^/]+)/(.*?) 歩(\d+)分')
df[['access2_route', 'access2_station', 'access2_time']] = df['access2'].str.extract(r'([^/]+)/(.*?) 歩(\d+)分')
df[['access3_route', 'access3_station', 'access3_time']] = df['access3'].str.extract(r'([^/]+)/(.*?) 歩(\d+)分')

# 不要になったaccess列を削除
df.drop(['access1','access2','access3'], axis=1, inplace=True)

#age列の処理：”築●年”を削除し、数値に変換
def convert_age(age_data):
    if '新築' in age_data:
        return 0
    else:
        return int(age_data.replace('築', '').replace('年', ''))
df['age'] = df['age'].apply(convert_age)

#height列の処理："階建"を削除し、数値に変換
df['height'] = pd.to_numeric(df['height'].str.rstrip('階建'),errors='coerce').fillna(0)

#floor列の処理："階"を削除し、数値に変換
df['floor'] = pd.to_numeric(df['floor'].str.rstrip('階'),errors='coerce').fillna(0)

#fee列の処理："万円"を削除し、数値に変換
df['fee'] = pd.to_numeric(df['fee'].str.rstrip('万円'),errors='coerce').fillna(0)

#management_fee列の処理："円"を削除し、数値に変換
df['management_fee'] = pd.to_numeric(df['management_fee'].str.rstrip('円'),errors='coerce').fillna(0)

# deposit列の処理: "万円" を削除し、数値に変換
df['deposit'] = pd.to_numeric(df['deposit'].str.rstrip('万円'), errors='coerce').fillna(0)

#gratuity列の処理："万円"を削除し、数値に変換
df['gratuity'] = pd.to_numeric(df['gratuity'].str.rstrip('万円'), errors='coerce').fillna(0)

#menseki列の処理："m2"を削除し、数値に変換
df['menseki'] = pd.to_numeric(df['menseki'].str.rstrip('m2'), errors='coerce').fillna(0)
```
